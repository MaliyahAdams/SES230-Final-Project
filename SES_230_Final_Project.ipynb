```python
# 1 Original
# Differential Equations

import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import odeint

# T an G are T_inf^-1 and T_inc^-1 (1/days) (respect.)!!
# N is the total population number

# SEIR model differential equations:
def SEIR(y, t, N, R_t, T, G):
    S, E, I, R = y
    beta = R_t / T
    dSdt = -beta * S * I / N
    dEdt = beta * S * I / N - E / T
    dIdt = E / T - I / G
    dRdt = I / G
    return np.array([dSdt, dEdt, dIdt, dRdt])

# initial conditions
I0 = 1
E0 = 0
R0 = 0
S0 = 1000 - I0 - E0 - R0
#variable
R_t = 2.2
T = 2.9
G = 10.0 #the lower it is, the flatter the curve: the less amount of time it takes for symptoms to show, the flatter the curve

t  = np.linspace(0, 100, 100)
y0 = np.array([S0,0,I0,R0]) #0,1,2,3 respect.

soln = odeint(SEIR, y0, t, args=(1000, R_t, T, G))

plt.figure(figsize = (10,8))
plt.plot(t, soln[:,0], 'r', label='Susceptible')
plt.plot(t, soln[:,1], 'g', label='Exposed')
plt.plot(t, soln[:,2], 'b', label='Infected')
plt.plot(t, soln[:,3], 'y', label='Recovered')
plt.xlabel('Days')
plt.ylabel('Population')
plt.title('SEIR Model')
plt.legend()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import odeint

# Simulitis model differential equations:
def simulitis(y, t, beta, T):
    S, I, R = y
    dSdt = -beta * S * I
    dIdt = beta * S * I - T * I
    dRdt = T * I
    return np.array([dSdt, dIdt, dRdt])

#initial conditions
I0 = 1
R0 = 0
S0 = 1000 - I0 - R0
#variables
beta = 0.2
T = 0.1


t = np.linspace(0, 100, 100)
y0 = np.array([S0, I0, R0]) #0,1,2 repect.

soln = odeint(simulitis, y0, t, args=(beta, T))

plt.figure(figsize=(10, 8))
plt.plot(t, soln[:, 0], 'r', label='Susceptible')
plt.plot(t, soln[:, 1], 'b', label='Infected')
plt.plot(t, soln[:, 2], 'y', label='Recovered')
plt.xlabel('Days')
plt.ylabel('Population')
plt.title('Simulitis Model')
plt.legend()
plt.show()
```
